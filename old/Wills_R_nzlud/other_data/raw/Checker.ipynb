```python
import pandas as pd

# Initialize an empty DataFrame for combined data
combined_data = pd.DataFrame()

# Loop through the range of your files
for i in range(48):  # 48 is exclusive, so it goes from 0 to 47
    file_name = f"output{i}_fin.xls"  # Construct file name
    data = pd.read_excel(file_name)  # Read the Excel file
    combined_data = pd.concat([combined_data, data], ignore_index=True)  # Append data

# Optionally, save the combined DataFrame to a new Excel file
combined_data.to_excel("combined_output_fin.xlsx", index=False)
```

```python
combined_data.to_csv("Mleczko_data.csv", index=False)
```
