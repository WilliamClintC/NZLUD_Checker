```python
import scipy
import scipy.cluster.hierarchy as sch

def cluster_corr(corr_array, inplace=False):
    """
    Rearranges the correlation matrix, corr_array, so that groups of highly 
    correlated variables are next to eachother 
    
    Parameters
    ----------
    corr_array : pandas.DataFrame or numpy.ndarray
        a NxN correlation matrix 
        
    Returns
    -------
    pandas.DataFrame or numpy.ndarray
        a NxN correlation matrix with the columns and rows rearranged
    """
    pairwise_distances = sch.distance.pdist(corr_array)
    linkage = sch.linkage(pairwise_distances, method='complete')
    cluster_distance_threshold = pairwise_distances.max()/2
    idx_to_cluster_array = sch.fcluster(linkage, cluster_distance_threshold, 
                                        criterion='distance')
    idx = np.argsort(idx_to_cluster_array)
    
    if not inplace:
        corr_array = corr_array.copy()
    
    if isinstance(corr_array, pd.DataFrame):
        return corr_array.iloc[idx, :].T.iloc[idx, :]
    return corr_array[idx, :][:, idx]

import numpy as np 
plt.figure(figsize=(12, 10))
sns.heatmap(cluster_corr(combined_df.corr()), annot=True, cmap='coolwarm', linewidths=0.5, vmin=-1, vmax=1, annot_kws={"size": 5})
plt.title('Cluster Correlation Full')
plt.show()
```

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
from geopy.geocoders import OpenCage
from geopy.exc import GeocoderTimedOut, GeocoderUnavailable
import time

# Path to your shapefile
shapefile_path = r'C:\Users\clint\Downloads\cb_2018_us_cbsa_500k\cb_2018_us_cbsa_500k.shp'

# Load the shapefile
gdf = gpd.read_file(shapefile_path)

# Initialize geolocator with OpenCage
api_key = 'ce393c93c9e94600b80afeb1b6cd06e7'  # Replace with your OpenCage API key
geolocator = OpenCage(api_key, timeout=10)  # Increase timeout duration

# Function to get coordinates with retry logic
def get_coordinates(place, state, retries=3):
    for _ in range(retries):
        try:
            location = geolocator.geocode(f"{place}, {state}")
            if location:
                return location.latitude, location.longitude
        except (GeocoderTimedOut, GeocoderUnavailable):
            time.sleep(1)  # Wait before retrying
    return None, None  # Return None if all retries fail

# Assuming combined_df is already defined and contains 'place' and 'statename' columns
combined_df['coordinates'] = combined_df.apply(lambda row: get_coordinates(row['place'], row['statename']), axis=1)
combined_df[['latitude', 'longitude']] = pd.DataFrame(combined_df['coordinates'].tolist(), index=combined_df.index)

# Plot the map
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
gdf.plot(ax=ax, color='white', edgecolor='black')

# Add points for each city
for idx, row in combined_df.iterrows():
    if pd.notnull(row['latitude']) and pd.notnull(row['longitude']):
        ax.plot(row['longitude'], row['latitude'], marker='o', color='red', markersize=5)

plt.show()
```
