```python
import pandas as pd

df = pd.read_csv('processed_combined_text_collector_aug7.csv')
```

```python
df = df[df['Num_Matches_Found'] == 1]
df
```

```python
df['Weight Result'] = df['Weight Result'].str.replace(r'[\[\]]', '', regex=True)
df
```

```python
value_counts_df = df['Initial_Match'].value_counts().reset_index()
value_counts_df.columns = ['Initial_Match', 'Count']
value_counts_df
```

```python
df['Weight Result'] = df['Weight Result'].astype(int)
df
```

```python


# Print the unique values in the 'Initial_Match' and 'Weight Result' columns
print(df['Initial_Match'].unique())
print(df['Weight Result'].unique())

# Print the data types of the columns
print(df.dtypes)
```

```python
# List of terms to check
terms = ['growth management', 'growth rate', 'growth control', 'minimum additional lot area']

# Initialize an empty list to store the results
results = []

# Loop through each term and count the successful matches
for term in terms:
    count = df[(df['Initial_Match'] == term) & (df['Weight Result'] == 1)].shape[0]
    results.append({'Initial_Match': term, 'Count': count})

# Create the successful_match_df DataFrame
successful_match_df = pd.DataFrame(results)

# Display the DataFrame
successful_match_df
```

```python

import matplotlib.pyplot as plt

# Create the first DataFrame
df1 = value_counts_df

# Create the second DataFrame
df2 = successful_match_df

# Merge the DataFrames on the 'Initial_Match' column
merged_df = pd.merge(df1, df2, on='Initial_Match', suffixes=('_df1', '_df2'))
merged_df
```

```python
merged_df['Ratio'] = merged_df['Count_df2'] / merged_df['Count_df1']*100
merged_df
```

```python
# Plot the data
fig, ax = plt.subplots()

# Plot the counts from the first DataFrame
ax.bar(merged_df['Initial_Match'], merged_df['Count_df1'], label='Initial Match (0 or 1)', alpha=0.7)

# Plot the counts from the second DataFrame
ax.bar(merged_df['Initial_Match'], merged_df['Count_df2'], label='Succesful Match (1)', alpha=0.7)

# Add labels and title
ax.set_xlabel('Initial Match')
ax.set_ylabel('Count')
ax.set_title('Initial Match')
ax.legend()

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```
