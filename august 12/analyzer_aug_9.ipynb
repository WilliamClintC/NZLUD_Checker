```python
import pandas as pd

df = pd.read_csv('processed_combined_text_collector_aug7.csv')
df.head()
```

```python
df_filtered = df[df['Num_Matches_Found'] != 0]
df_filtered
```

```python
from collections import Counter
import re
import pandas as pd

# Assuming df is your DataFrame and +-250_Text is the column name
# Combine all text entries into a single string
all_text = ' '.join(df['+-250_Text'].dropna().tolist())

# Tokenize the string into individual words
words = re.findall(r'\b\w+\b', all_text.lower())
```

```python

# Count the frequency of each word
word_counts = Counter(words)
#word_counts
word_counts_list = list(word_counts.items())

# Step 4: Create a DataFrame from the list of tuples
df = pd.DataFrame(word_counts_list, columns=['Word', 'Count'])
df
```

```python
df_sorted = df.sort_values('Count', ascending=False).reset_index(drop=True)
df_sorted
```

```python

from nltk.corpus import stopwords
import nltk

# Download stop words
nltk.download('stopwords')
df=df_sorted

# List of stop words
stop_words = set(stopwords.words('english'))

# Filter out stop words
df_filtered = df[~df['Word'].isin(stop_words)]

df_filtered
```

```python
# Assuming df_filtered is a pandas DataFrame
df_filtered.to_csv('word_count.csv', index=False)
```
